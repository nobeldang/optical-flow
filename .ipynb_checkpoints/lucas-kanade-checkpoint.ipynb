```python
window_size = 5
```

```python
def getANeighbourhood(i, j, Ix, Iy, window_size):
    st = window_size//2
    A = np.zeros((2, 2))
    Ix_window = Ix[i - st: i + st][j - st, j + st]
    Iy_window = Iy[i - st: i + st][j - st, j + st]
    Ixy_window = Ix_window*Iy_window
    sIx = np.sum(Ix_window)
    sIy = np.sum(Iy_window)
    sIxy = np.sum(Ixy_window)
    A[0][0] = sIx**2
    A[1][1] = sIy**2
    A[0][1] = sIxy
    A[1][0] = sIxy
    
    return A
```

```python
def getbNeighbourhood(i, j, Ix, Iy, It, window_size):
    st = window_size//2
    b = np.zeros((2, 1))
    It_window = It[i - st: i + st][j - st, j + st]
    Ix_window = Ix[i - st: i + st][j - st, j + st]
    Iy_window = Iy[i - st: i + st][j - st, j + st]
    b[0] = np.sum(It_window*Ix_window)
    b[1] = np.sum(It_window*Iy_window)
    return b
```

```python
def lucasKanade(img1, img2, Ix, Iy, It, window_size = 5):
    st = window_size//2
    u = np.zeros((img1.shape))
    v = np.zeros((img2.shape))
    for i in range(st, img1.shape[0] - st):
        for j in range(st, img1.shape[1] - st):
            A = getANeighbourhood(i, j, Ix, Iy, window_size)   ### this is A'A
            B = getbNeighbourhood(i, j, Ix, Iy, It, window_size)   ### this is B = A'b
            velocityVec = np.linalg.inv(A)*B   ### soloution: inv(A'A)*B
            print(velVector)
            u[i, j] = velocityVec[0]
            v[i, j] = velocityVec[1]
    return u, v
```

```python
vel = lucasKanade(img1, img2, Ix, Iy, It, 5)
```
